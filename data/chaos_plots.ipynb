```python
import os
import numpy as np
def read_bin(filename):
    sim = rebound.SimulationArchive(filename)
    try:
        earth = archive[0].particles['earth']
        small_planets_flag = True
    except:
        small_planets_flag = False
     
    clones_flag = False
    if small_planets_flag and len(sim[-1].particles) > 10:
        clones_flag = True
        numclones = len(sim[-1].particles)-10
        objnum = 9
    elif not small_planets_flag and len(sim[-1].particles) > 6:
        clones_flag = True
        numclones = len(sim[-1].particles)-6
        #print('nuclones',numclones)
        objnum = 5
    else:
        numclones = 0
        
    a = np.zeros((numclones+1,len(sim)))    
    e = np.zeros((numclones+1,len(sim)))    
    I = np.zeros((numclones+1,len(sim)))
    for i in range(len(sim)):
        for j in range(numclones+1):
            a[j][i] = sim[i].particles[objnum+j].a
            e[j][i] = sim[i].particles[objnum+j].e
            I[j][i] = sim[i].particles[objnum+j].inc
    return a,e,I
    
def calc_chaos(objname,objtype):
    #print(objname)
    file = '../data/' + objtype + '/' + str(objname)

    sim = rebound.SimulationArchive(file + "/archive_chaos_long.bin")
    if os.path.exists(file + "/archive_chaos_long.bin"):
        try:
            earth = archive[0].particles['earth']
            small_planets_flag = True
        except:
            small_planets_flag = False
        
        clones_flag = False
        if small_planets_flag and len(sim[-1].particles) > 10:
            clones_flag = True
            numclones = len(sim[-1].particles)-10
            objnum = 9
        elif not small_planets_flag and len(sim[-1].particles) > 6:
            clones_flag = True
            numclones = len(sim[-1].particles)-6
            #print('nuclones',numclones)
            objnum = 5
        else:
            numclones = 0
    
        a,e,I = read_bin(file + "/archive_chaos_long.bin")
        #sb = long_sim[-1].particles[5]
        #print(sb)
            
        a_diff = np.zeros((numclones,len(a[0])))       
        e_diff = np.zeros((numclones,len(e[0])))       
        I_diff = np.zeros((numclones,len(I[0])))        
        
        for i in range(numclones):
            #print(sb.a,sb.e,sb.inc)           
            #if a[0][-1] < 1:
            #    a_diff[i] = [0,0,0]
                #CloneDist = [0,0,0]
            #    continue
            #print(len(a),objnum,numclones,len(sim[-1].particles))
            a_diff[i] = ((a[0] - a[i+1])/a[0])**2
            e_diff[i] = ((e[0] - e[i+1]))**2
            I_diff[i] = ((np.sin(I[0]) - np.sin(I[i+1])))**2
        #print(sb.a,sb.e,sb.inc)
        #print(long_sim[-1].particles[objnum+i].a,long_sim[-1].particles[objnum+i].e,long_sim[-1].particles[objnum+i].inc)
        #print(a_diff)
        CloneDist = np.sqrt(np.nanmean(np.array([a_diff,e_diff,I_diff]),axis=0))
    else:
        CloneDist = np.zeros((3,len(a)))
            
    return CloneDist[0],CloneDist[1],CloneDist[2]
```
