```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def hcm(a,rms_a,rms_e,rms_i):
    F = 365*24*60*60/1.496e11
    return (2*np.pi*1/F*np.sqrt((5/4*(rms_a/a)**2+2*rms_e**2+2*rms_i**2)/a))

df = pd.read_csv('NesvornyAst_prop_elem_multi_5win_helio_1000.csv')
df = pd.read_csv('NesvornyAst_prop_elem_multi_5win_helio_10000.csv')
df = pd.read_csv('NesvornyAst_prop_elem_multi_1000.csv')
#df = pd.read_csv('NesvornyAst_prop_elem_multi_10000.csv')
#df = pd.read_csv('NesvornyAst_prop_elem_multi_10000_rms.csv')

hcm1 = hcm(df['PropSMA'],df['RMS_err_a'],df['RMS_err_e'],df['RMS_err_sinI'])
bins = 10**(np.arange(-2.0,3.0,1/20))
plt.hist(hcm1,bins=bins)
plt.xscale('log')
plt.title('Nesvorny')
plt.show()

ind = np.where(hcm1 > 100)[0]
#print(df.iloc[ind])
#print(hcm1.iloc[ind])
```
