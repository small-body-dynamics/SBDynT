```python
import numpy as np

def Rtot(Fa,Fb,Ra,Rb,d):
    Ftot = Fa+Fb
    return np.sqrt((Fa*Ra**2+Fb*Rb**2+Fa*Fb*d**2/Ftot)/Ftot)

Ra = 0.42

Rb = Ra*0.1375 #160 km object
Rb = Ra*0.1255 #146 km object
Rb = Ra*0.127257 #148 km object

SA_a = np.pi*Ra**2
SA_b = np.pi*Rb**2

SAtot = SA_a+SA_b

Fa = SA_a/SAtot
Fb = SA_b/SAtot
d = Ra*10.7

print('Radius at 12418 km separation:',Rtot(Fa,Fb,Ra,Rb,d))

d = 15125/1163*Ra

print('Radius at 15125 km separation:',Rtot(Fa,Fb,Ra,Rb,d))

d = 9710/1163*Ra

print('Radius at 9710 km (periapse) separation:',Rtot(Fa,Fb,Ra,Rb,d))
```
