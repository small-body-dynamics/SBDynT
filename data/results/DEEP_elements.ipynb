```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('DEEP_prop_elem.csv')
print(data)
```

```python


plt.scatter(data['ObsEcc'],data['PropEcc'])
plt.show()
plt.scatter(data['ObsSin(Inc)'],data['PropSin(Inc)'])
plt.show()
plt.scatter(data['ObsEcc'],data['ObsSin(Inc)'])
plt.show()
plt.scatter(data['PropEcc'],data['PropSin(Inc)'])
plt.show()

ind = np.where(data['PropSMA'] > 0)[0]
wroind = np.where(data['PropSMA'] < 0)[0]
print(wroind)

plt.scatter(data['PropSMA'][ind],data['PropSin(Inc)'][ind])
plt.show()
plt.scatter(data['PropSMA'][ind],data['PropEcc'][ind])
plt.show()
```

```python
TNOs = pd.read_csv('TNOs_prop_elem.csv')

indTNO1 = np.where(TNOs['PropSMA'] > 20)[0]
indTNO2 = np.where(TNOs['PropSMA'] < 100)[0]
indTNO3 = np.where(TNOs['PropEcc'] < 1)[0]

indTNO = np.intersect1d(indTNO1, indTNO2)
indTNO = np.intersect1d(indTNO, indTNO3)

indDEEP = np.where(data['PropSMA'] > 0)[0]

fig,ax = plt.subplots(2,figsize=(7,8),sharex=True)
fig.subplots_adjust(hspace=0)
ax[0].scatter(TNOs['PropSMA'][indTNO],TNOs['PropEcc'][indTNO],s=24,alpha=0.5)
ax[0].scatter(data['PropSMA'][indDEEP],data['PropEcc'][indDEEP],s=24)
#ax[0].set_xlabel('Proper Semi-Major Axis')
ax[0].set_ylabel('Proper Eccentricity')
ax[0].set_title('Proper Elements from Smotherman et al. 2023')

ax[1].scatter(TNOs['PropSMA'][indTNO],TNOs['PropSin(Inc)'][indTNO],s=24,alpha=0.5)
ax[1].scatter(data['PropSMA'][indDEEP],data['PropSin(Inc)'][indDEEP],s=24)
ax[1].set_xlabel('Proper Semi-Major Axis')
ax[1].set_ylabel('Proper Sin(Inclination)')
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(TNOs['PropEcc'][indTNO],TNOs['PropSin(Inc)'][indTNO],s=24,alpha=0.5)
plt.scatter(data['PropEcc'][indDEEP],data['PropSin(Inc)'][indDEEP],s=24)
plt.xlabel('Proper Eccentricity')
plt.ylabel('Proper Sin(Inclination)')
plt.show()
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'},figsize=(8,8))
ax.scatter(np.arcsin(TNOs['PropSin(Inc)'][indTNO]), TNOs['PropSMA'][indTNO],s=15,alpha=0.5)
ax.scatter(np.arcsin(data['PropSin(Inc)'][indDEEP]), data['PropSMA'][indDEEP],s=25,c='r')
ax.set_rmin(32)
ax.set_rmax(66)
ax.set_ylabel('Proper Sin(Inclination)')
ax.set_xlabel('Prop Semi-Major Axis')
ax.set_xlim(0,np.pi/4)
```
