```python
def rms(x, axis=None):
    return np.sqrt(np.nanmean(x**2, axis=axis))

TNOs_err = TNOs[TNOs.columns[-9:]]
DESs_err = DES[DES.columns[-9:]]
datas_err = data[data.columns[-9:]]
TNO_err = np.zeros((len(TNOs_err),3))
DES_err = np.zeros((len(DESs_err),3))
data_err = np.zeros((len(datas_err),3))
import json
var = "[[2149, 2349], [4536, 4844], [3539, 3779]]"
var = json.loads(var)
for i in range(len(TNOs_err)):
    var = np.zeros((len(TNOs_err.iloc[i]),3))
    for j in range(len(TNOs_err.iloc[i])):
        #print(TNOs_err.iloc[i][j])
        dat =  json.loads(TNOs_err.iloc[i][j])
        #print(dat)
        var[j] = dat
    prop = [TNOs['PropEcc'].iloc[i],TNOs['PropSin(Inc)'].iloc[i],TNOs['PropSMA'].iloc[i]]
    TNO_err[i] = rms(prop-var,axis=0)
    
for i in range(len(DESs_err)):
    var = np.zeros((len(DESs_err.iloc[i]),3))
    for j in range(len(DESs_err.iloc[i])):
        #print(TNOs_err.iloc[i][j])
        dat =  json.loads(DESs_err.iloc[i][j])
        #print(dat)
        var[j] = dat
    
    prop = [DES['PropEcc'].iloc[i],DES['PropSin(Inc)'].iloc[i],DES['PropSMA'].iloc[i]]
    DES_err[i] = rms(prop-var,axis=0)
    
for i in range(len(datas_err)):
    var = np.zeros((len(datas_err.iloc[i]),3))
    for j in range(len(datas_err.iloc[i])):
        #print(TNOs_err.iloc[i][j])
        dat =  json.loads(datas_err.iloc[i][j])
        #print(dat)
        var[j] = dat
    
    prop = [data['PropEcc'].iloc[i],data['PropSin(Inc)'].iloc[i],data['PropSMA'].iloc[i]]
    data_err[i] = rms(prop-var,axis=0)
    #print(TNO_err[i])
#print(TNOs[TNOs.columns[-9:]])
print(TNOs_err)
```
