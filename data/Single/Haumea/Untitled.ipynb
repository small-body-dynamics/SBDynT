```python
import rebound
import numpy as np
sim= rebound.Simulationarchive('archive.bin')
sim2 = rebound.Simulationarchive('../../TNOs/136108/archive.bin')
#sim2 = rebound.Simulationarchive('../2003 EL61/archive.bin')
sim2 = rebound.Simulationarchive('../136108/archive.bin')
print(sim[-1].t)
print(sim2[-1].t)
print(sim[-1].dt,sim2[-1].dt)
```

```python
n = len(sim)
n2 = len(sim2)
a = np.zeros(n)
e = np.zeros(n)
inc = np.zeros(n)
a2 = np.zeros(n2)
e2 = np.zeros(n2)
inc2 = np.zeros(n2)
for i in range(n):
    a[i] = sim[i].particles[-1].a
    e[i] = sim[i].particles[-1].e
    inc[i] = sim[i].particles[-1].inc
for i in range(n2):
    a2[i] = sim2[i].particles[-1].a
    e2[i] = sim2[i].particles[-1].e
    inc2[i] = sim2[i].particles[-1].inc
```

```python
import matplotlib.pyplot as plt
plt.scatter(range(n),a)
plt.show()
plt.scatter(range(n2),a2)
plt.show()
plt.scatter(range(n),e)
plt.show()
plt.scatter(range(n2),e2)
plt.show()
plt.scatter(range(n),inc)
plt.show()
plt.scatter(range(n2),inc2)
plt.show()
```
