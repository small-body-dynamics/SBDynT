```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
kB = 8.617333262145e-5  # Boltzmann constant in eV/K
EF0 = 1.0  # Fermi energy at T = 0 in eV

# Function to calculate Fermi energy at a given temperature
def fermi_energy(T):
    return EF0 * np.exp(-EF0 / (2 * kB * T))

# Function to calculate Fermi-Dirac distribution
def fermi_dirac(E, T):
    return 1 / (np.exp((E - fermi_energy(T)) / (kB * T)) + 1)

# Energy range
E_range = np.linspace(0, 2*EF0, 1000)

# Temperature range
T_range = np.linspace(0.1, 5, 20)

# Plotting
plt.figure(figsize=(10, 6))
for T in T_range:
    plt.plot(E_range, fermi_dirac(E_range, T), label=f'T = {T:.2f} K')

plt.axvline(x=EF0, color='k', linestyle='--', label='Fermi energy at T=0')
plt.xlabel('Energy (eV)')
plt.ylabel('Probability')
plt.title('Electron Degeneracy as a Function of Temperature')
plt.legend()
plt.grid(True)
plt.show()
```
