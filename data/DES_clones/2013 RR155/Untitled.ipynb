```python
import numpy as np
import matplotlib.pyplot as plt
sim = rebound.SimulationArchive('archive.bin')

length = len(sim)

a = np.zeros(length)
e = np.zeros(length)
inc = np.zeros(length)
for i in range(length):
    a[i] = sim[i].particles[-1].a
    e[i] = sim[i].particles[-1].e
    inc[i] = sim[i].particles[-1].inc
    
plt.scatter(np.arange(length),a)
plt.show()
plt.scatter(np.arange(length),e)
plt.show()
plt.scatter(np.arange(length),inc)
plt.show()

print(init.particles[-1].a,init.particles[-1].e,init.particles[-1].inc)
```

```python
ffte = np.fft.rfft(e)

plt.scatter(np.arange(len(ffte))[2:],ffte[2:])
plt.yscale('log')
plt.xscale('log')
plt.xlim(1e2,1e4)
```
