```python
import scipy.signal as signal
import numpy as np

t = np.arange(0,40*np.pi-10,0.01)
freq = np.logspace(1e-4,1,len(t))
h = np.sin(t)+2*np.cos(2*t)+4*np.sin(34*t)
pgram = signal.lombscargle(t, h, freq, normalize=True)
from astropy.timeseries import LombScargle
frequency, power = LombScargle(t, h).autopower()
```

```python
plt.scatter(freq,pgram)
plt.xscale('log')
plt.yscale('log')
plt.show()
plt.scatter(frequency,power)
#plt.xscale('log')
plt.yscale('log')
plt.show()

print(len(freq),len(np.fft.rfft(h)))
plt.scatter(np.fft.ifft(np.fft.rfft(freq[:])),np.fft.rfft(h[:]))
plt.xscale('log')
#plt.yscale('log')
plt.show()
```
