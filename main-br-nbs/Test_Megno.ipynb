```python
#print(series)
dx = np.array([series['x']-series2['x'],series['y']-series2['y'],series['z']-series2['z']])
dv = np.array([series['vx']-series2['vx'],series['vy']-series2['vy'],series['vz']-series2['vz']])
v = np.array([series['vx']-series2['vx'],series['vy']-series2['vy'],series['vz']-series2['vz']])
dx_t = (dx[:,1])
dx_0 = (dx[:,0])

print(dx,dx[:,0],dx[0,:])

print(np.log(np.linalg.norm(dx_t/dx_0)))
print(series)
```

```python
def Y(N):
    Yn = 0
    for k in range(1,N):
        Yn = Yn + np.log(np.linalg.norm(dv[:,k])/np.linalg.norm(dv[:,k-1]))*k**2
    return 3*Yn

def Ybar(N):
    temp = 0
    for i in range(1,N):
        temp = temp + Y(i)
    return 1/N**3*temp

def Ycar(N):
    return (4*Ybar(N)-2)/N
MEGNO = np.zeros(10)
for i in range(1,10):
    MEGNO[i] = Ycar(i)
print(MEGNO)
```

```python
dt = 1000
def deltad_delta(t):
    deltad = 0
    delta2 = 0
    
    deltad = deltad + series['vx'][t]*series['x'][t]
    deltad = deltad + series['vy'][t]*series['y'][t]
    deltad = deltad + series['vz'][t]*series['z'][t]
    deltad = deltad + series['ax'][t]*series['vx'][t]
    deltad = deltad + series['ay'][t]*series['vy'][t]
    deltad = deltad + series['az'][t]*series['vz'][t]
    
    delta2 = delta2 + series['x'][t]**2+series['y'][t]**2+series['z'][t]**2
    delta2 = delta2 + series['vx'][t]**2+series['vy'][t]**2+series['vz'][t]**2
        
    return deltad/delta2

def dY(t):
    return 2*t*deltad_delta(t)

def Ys(t):
    if t == 0:
        return 0
    return (Ys(t-1)+dY(t))

def Y(t):
    return Ys(t)/t

def Yss(t):
    if t ==0:
        return 0
    return Yss(t-1)+Y(t)

print(Yss(200))
```
